```python
import pandas as pd

# Read the Jupyter Notebook file as a pandas dataframe
df = pd.read_csv('/Users/mridul/Desktop/Physics Research Summer/CV-Analyzer/New Data/Data Files/aavsodata_AH Her.csv')

Data_B = df[df['Band'] == 'B']
Data_B_filteres = Data_B[['JD', 'Magnitude', 'Band']]
Data_B_filteres['Magnitude'] = Data_B_filteres['Magnitude'].astype(float)
Data_B_filteres['JD'] = Data_B_filteres['JD'] - 2458880

```

```python
Data_B_filteres.to_csv('random.csv', index=False)
```

```python
plt.figure()
Data_B_filteres.plot(x = 'JD', y= 'Magnitude', marker = '.', linewidth=0.3)
plt.show()
```

```python
from astropy.timeseries import LombScargle

frequency, power = LombScargle(Data_B_filteres['JD'], Data_B_filteres['Magnitude']).autopower(maximum_frequency=2)

plt.figure()
plt.plot(frequency, power)

```

```python
import numpy as np
rand = np.random.default_rng(42)
t = 100 * rand.random(100)
y = np.sin(2 * np.pi * t) + 0.1 * rand.standard_normal(100)
```

```python
plt.figure()
plt.plot(t, y, 'o', linewidth=0.1)
```

```python
from astropy.timeseries import LombScargle
frequency, power = LombScargle(t, y).autopower(maximum_frequency=30)
```

```python
import matplotlib.pyplot as plt  
plt.figure()
plt.plot(frequency, power)   
plt.show()    
```
